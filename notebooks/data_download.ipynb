This cell sets up the Kaggle API credentials by copying the `kaggle.json` file from Google Drive to the correct directory and setting appropriate file permissions.

```python
import os

# Create the .kaggle directory if it doesn't exist
os.makedirs(os.path.expanduser('~/.kaggle'), exist_ok=True)

# Define the source path for your kaggle.json file
source_kaggle_json = '/content/drive/MyDrive/פרויקט גמר RMOT/Code Pipeline Example/kaggle_API/kaggle.json'

# Define the target directory for the kaggle.json file
target_kaggle_dir = os.path.expanduser('~/.kaggle')

# Copy the kaggle.json file into the .kaggle directory
# Adding a trailing slash to target_kaggle_dir explicitly tells cp it's a directory
!cp "{source_kaggle_json}" "{target_kaggle_dir}/"

# Set permissions for the kaggle.json file (which will now be inside ~/.kaggle/)
!chmod 600 "{os.path.join(target_kaggle_dir, 'kaggle.json')}"
```
